```python
import numpy as np
import matplotlib.pyplot as plt

import jieba
from wordcloud import WordCloud
import PIL.Image as Image
```

```python
# 载入文本数据
with open('qixi.txt', 'r',encoding='UTF-8') as f:
#     print(f.read())
    txt = f.read()

print(txt)
```

```python
txt2 = " ".join(jieba.cut(txt))
txt2
```

```python
# 载入停用词文件
with open('stopword.txt', 'r',encoding='UTF-8') as f:
#     print(f.read())
    s = f.read()
stopword = s.split('\n')

stopword
```

```python

  #2.图片遮罩层
mask_pic = np.array(Image.open("timg.jpg"))

wordcloud = WordCloud(background_color ="white", font_path="arial unicode ms.ttf",stopwords = stopword,mask = mask_pic).generate(txt)
image=wordcloud.to_image()
image.show()

plt.figure(2,dpi=300)
plt.imshow(wordcloud, interpolation='bilinear') # 绘制数据内的图片,双线性插值绘图
plt.axis("off") # 去掉坐标轴
plt.show()
```
